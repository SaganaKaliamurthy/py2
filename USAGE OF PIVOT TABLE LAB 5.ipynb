```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
titanic=sns.load_dataset('titanic')
titanic
```

```python
# pivot table to view show male and female passenger according to their class:
```

```python
titanic.pivot_table('survived',index='sex',columns='class')
```

```python
# row wise
```

```python
age=pd.cut(titanic['age'],[0,18,80])
titanic.pivot_table('survived',['sex','age'],'class')
```

```python
# column wise
```

```python
fare=pd.qcut(titanic['fare'],2)
titanic.pivot_table('survived',['sex',age],[fare,'class'])
```

```python
titanic.pivot_table(index='sex',columns='class',
                   aggfunc={'survived':sum,'fare':'mean'})
```

```python
# table to show only survived passengers list:
```

```python
titanic.pivot_table('survived',index='sex',columns='class',margins=True)
```

```python
# importing birth dates dataset:
```

```python
titanic.pivot_table(index='sex',columns='class',aggfunc={'survived':'count','fare':'mean'})
```

```python
age=pd.cut(titanic['age'],[0,18,80])
titanic.pivot_table(values="fare",index=["sex",age],columns="class")
```

```python
fare=pd.cut(titanic['age'],[0,18,80])
titanic.pivot_table('survived',['sex',age],'embark_town')
```
